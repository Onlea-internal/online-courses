# Types of Correlations
## Activity for Module 5 Unit 2 

```python
# linear
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(1,100,100)
y = x + 3
plt.plot(x, y)
plt.show()
```

```python
# linear with negative slope
y = -x + 3
plt.plot(x, y)
plt.show()
```

```python
# exponential
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
# inverse
y = 1 / x
plt.plot(x, y)
plt.show()
```
